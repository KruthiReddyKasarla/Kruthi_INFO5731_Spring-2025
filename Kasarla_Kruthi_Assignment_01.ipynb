# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import math

# volume of a hemisphere
def calculate_hemisphere_volume(r):
    return (2 / 3) * math.pi * r**3

# even numbers from 1 to 20
for r in range(2, 21, 2):
    V = calculate_hemisphere_volume(r)
    print(f"Radius: {r}, Volume: {V:.2f}")


```

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
import string

# (a)
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# (b)
def sentence_to_word_list(sentence):
    return sentence.split()

sentence1_words = sentence_to_word_list(sentence1)
sentence2_words = sentence_to_word_list(sentence2)

print("\nSentence 1 as a list of words:", sentence1_words)
print("Sentence 2 as a list of words:", sentence2_words)

# (c)
common_words = list(set(sentence1_words) & set(sentence2_words))
print("\nWords that occur in both sentences:", common_words)

# (d)
def remove_punctuation(sentence):
    return sentence.translate(str.maketrans("", "", string.punctuation))

sentence1_sanitized = remove_punctuation(sentence1)
sentence2_sanitized = remove_punctuation(sentence2)

print("\nSanitized Sentence 1:", sentence1_sanitized)
print("Sanitized Sentence 2:", sentence2_sanitized)

# (e)
def count_vowels(sentence):
    vowels = "aeiou"
    vowel_count = {v: sentence.lower().count(v) for v in vowels}
    return vowel_count

sentence1_vowels = count_vowels(sentence1)
sentence2_vowels = count_vowels(sentence2)

print("\nVowel counts in Sentence 1:", ', '.join(f"{v}:{count}" for v, count in sentence1_vowels.items()))
print("Vowel counts in Sentence 2:", ', '.join(f"{v}:{count}" for v, count in sentence2_vowels.items()))


```

# Question 4 (10 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
import math

# evaluate mathematical expressions
def evaluate_expression(expression):
    try:
        # using Python's eval() function
        result = eval(expression)
        return result
    except ZeroDivisionError:
        # division by zero
        return "Error: Division by zero is not allowed."
    except Exception as e:
        # other exceptions, invalid syntax
        return f"Error: {str(e)}"

# user interaction
def calculator():
    print("Simple Command-Line Calculator")
    print("Supported operations: +, -, *, /, ** (exponentiation), sqrt (square root), % (modulus)")
    print("Type 'exit' to quit the calculator.")

    while True:
        expression = input("\nEnter an expression: ").strip()

        if expression.lower() == 'exit':
            print("Exiting the calculator. Goodbye!")
            break

        # Check for square root operation
        if 'sqrt' in expression:
            try:
                number = float(expression.split('(')[1].split(')')[0])
                result = math.sqrt(number)
                print(f"Result: {result}")
                continue
            except Exception:
                print("Error: Invalid square root input.")
                continue

        # Evaluate mathematical expression
        result = evaluate_expression(expression)
        print(f"Result: {result}")

# Run calculator
if __name__ == "__main__":
    calculator()
```

```python
import csv
import os

# File name for storing inventory data
INVENTORY_FILE = "inventory.csv"

# Categories for validation
VALID_CATEGORIES = ["Electronics", "Furniture", "Groceries", "Clothing", "Books", "Toys"]

# Function to initialize the inventory if the file is missing or corrupted
def initialize_inventory():
    if not os.path.exists(INVENTORY_FILE):
        with open(INVENTORY_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Product ID", "Name", "Category", "Quantity", "Price"])
            print("Inventory file created.")

# Function to load inventory from CSV
def load_inventory():
    try:
        with open(INVENTORY_FILE, mode='r') as file:
            reader = csv.DictReader(file)
            inventory = [row for row in reader]
            return inventory
    except Exception as e:
        print(f"Error loading inventory: {e}")
        return []

# Function to save inventory back to CSV
def save_inventory(inventory):
    with open(INVENTORY_FILE, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Product ID", "Name", "Category", "Quantity", "Price"])
        for item in inventory:
            writer.writerow([item["Product ID"], item["Name"], item["Category"], item["Quantity"], item["Price"]])

# Function to add a product
def add_product(product_id, name, category, quantity, price):
    inventory = load_inventory()

    # Edge Case 1: Prevent duplicate product IDs
    if any(item["Product ID"] == product_id for item in inventory):
        print("Error: Product ID already exists.")
        return

    # Edge Case 4: Validate input data
    if int(quantity) < 0 or float(price) < 0:
        print("Error: Quantity and price cannot be negative.")
        return

    # Edge Case 10: Validate category
    if category not in VALID_CATEGORIES:
        print("Error: Invalid category.")
        return

    inventory.append({"Product ID": product_id, "Name": name, "Category": category, "Quantity": quantity, "Price": price})
    save_inventory(inventory)
    print(f"Product '{name}' added successfully.")

# Function to delete a product
def delete_product(product_id):
    inventory = load_inventory()

    # Edge Case 2: Deleting a non-existing product
    new_inventory = [item for item in inventory if item["Product ID"] != product_id]
    if len(new_inventory) == len(inventory):
        print("Error: Product not found.")
        return

    save_inventory(new_inventory)
    print(f"Product '{product_id}' deleted successfully.")

# Function to update product details
def update_product(product_id, new_name=None, new_category=None, new_quantity=None, new_price=None):
    inventory = load_inventory()

    # Edge Case 3: Updating a non-existing product
    found = False
    for item in inventory:
        if item["Product ID"] == product_id:
            found = True
            if new_name:
                item["Name"] = new_name
            if new_category:
                if new_category not in VALID_CATEGORIES:
                    print("Error: Invalid category.")
                    return
                item["Category"] = new_category
            if new_quantity is not None:
                if int(new_quantity) < 0:
                    print("Error: Quantity cannot be negative.")
                    return
                item["Quantity"] = new_quantity
            if new_price is not None:
                if float(new_price) < 0:
                    print("Error: Price cannot be negative.")
                    return
                item["Price"] = new_price
            break

    if not found:
        print("Error: Product not found.")
        return

    save_inventory(inventory)
    print(f"Product '{product_id}' updated successfully.")

# Function to search for a product
def search_product(product_id):
    inventory = load_inventory()

    # Edge Case 5: Searching for a non-existing product
    for item in inventory:
        if item["Product ID"] == product_id:
            print(f"Product Found: {item}")
            return
    print("Product not found.")

# Function to automate restocking when quantity is low
def restock_inventory(threshold=5, restock_amount=10):
    inventory = load_inventory()
    restocked = False

    # Edge Case 7: Restocking automation
    for item in inventory:
        if int(item["Quantity"]) < threshold:
            item["Quantity"] = str(int(item["Quantity"]) + restock_amount)
            restocked = True

    if restocked:
        save_inventory(inventory)
        print("Inventory restocked for low-stock items.")

# Function to generate reports
def generate_report():
    inventory = load_inventory()
    category_summary = {}

    print("\nInventory Report:")
    for item in inventory:
        print(f"{item['Name']} (Category: {item['Category']}): {item['Quantity']} units @ ${item['Price']} each")

        # Edge Case 9: Handling large inventories
        category_summary[item["Category"]] = category_summary.get(item["Category"], 0) + int(item["Quantity"])

    print("\nCategory Breakdown:")
    for category, total in category_summary.items():
        print(f"{category}: {total} items")

# Initialize the inventory system
initialize_inventory()

# Example Usage
add_product("101", "Laptop", "Electronics", "10", "999.99")
add_product("102", "Sofa", "Furniture", "5", "499.99")
add_product("103", "Apples", "Groceries", "30", "1.99")
delete_product("104")  # Edge Case: Deleting non-existing product
update_product("101", new_price="1099.99")  # Updating an existing product
search_product("102")  # Searching for a product
restock_inventory()  # Automating restocking
generate_report()  # Generating the final report
```

```python
import csv
import os

# File to store flight data
FLIGHT_FILE = "flights.csv"

# Initialize 50 sample flights (Flight ID, Destination, Seats Available, Price)
DEFAULT_FLIGHTS = [
    ["F101", "New York", 150, 350.0],
    ["F102", "Los Angeles", 200, 420.0],
    ["F103", "Chicago", 120, 290.0],
    ["F104", "Houston", 100, 310.0],
    ["F105", "Miami", 80, 280.0],
] + [[f"F10{i}", f"City{i}", 100 + (i % 50), 300 + (i % 50) * 2.5] for i in range(6, 56)]

# Function to initialize flight records
def initialize_flights():
    if not os.path.exists(FLIGHT_FILE):
        with open(FLIGHT_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Flight ID", "Destination", "Seats Available", "Price"])
            writer.writerows(DEFAULT_FLIGHTS)
        print("Flight data initialized.")

# Function to load flights from the CSV file
def load_flights():
    try:
        with open(FLIGHT_FILE, mode='r') as file:
            reader = csv.DictReader(file)
            return [row for row in reader]
    except Exception as e:
        print(f"Error loading flight data: {e}")
        return []

# Function to save flight data back to CSV
def save_flights(flights):
    with open(FLIGHT_FILE, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Flight ID", "Destination", "Seats Available", "Price"])
        for flight in flights:
            writer.writerow([flight["Flight ID"], flight["Destination"], flight["Seats Available"], flight["Price"]])

# Function to add a new flight
def add_flight(flight_id, destination, seats, price):
    flights = load_flights()

    # Edge Case 1: Prevent duplicate flight IDs
    if any(flight["Flight ID"] == flight_id for flight in flights):
        print("Error: Flight ID already exists.")
        return

    # Edge Case 5: Validate seat count and price
    if int(seats) <= 0 or float(price) <= 0:
        print("Error: Seats and price must be positive.")
        return

    flights.append({"Flight ID": flight_id, "Destination": destination, "Seats Available": str(seats), "Price": str(price)})
    save_flights(flights)
    print(f"Flight '{flight_id}' to {destination} added successfully.")

# Function to search for flights
def search_flight(flight_id):
    flights = load_flights()

    # Edge Case 2: Handle non-existent flight
    for flight in flights:
        if flight["Flight ID"] == flight_id:
            print(f"Flight Found: {flight}")
            return
    print("Flight not found.")

# Function to book seats
def book_seats(flight_id, num_seats):
    flights = load_flights()

    # Edge Case 3: Prevent overbooking
    for flight in flights:
        if flight["Flight ID"] == flight_id:
            available_seats = int(flight["Seats Available"])
            if num_seats > available_seats:
                print(f"Error: Only {available_seats} seats available.")
                return
            flight["Seats Available"] = str(available_seats - num_seats)
            save_flights(flights)
            print(f"Successfully booked {num_seats} seat(s) on flight {flight_id}.")
            return

    print("Error: Flight not found.")

# Function to cancel a reservation
def cancel_seats(flight_id, num_seats):
    flights = load_flights()

    # Edge Case 4: Prevent canceling unreserved seats
    for flight in flights:
        if flight["Flight ID"] == flight_id:
            available_seats = int(flight["Seats Available"])
            flight["Seats Available"] = str(available_seats + num_seats)
            save_flights(flights)
            print(f"Successfully canceled {num_seats} seat(s) on flight {flight_id}.")
            return

    print("Error: Flight not found.")

# Function to generate reports
def generate_report():
    flights = load_flights()
    busiest_flight = max(flights, key=lambda f: int(f["Seats Available"]))
    emptiest_flight = min(flights, key=lambda f: int(f["Seats Available"]))

    print("\n📊 Flight Report:")
    print(f"🚀 Most available seats: {busiest_flight['Flight ID']} ({busiest_flight['Seats Available']} seats)")
    print(f"🔴 Least available seats: {emptiest_flight['Flight ID']} ({emptiest_flight['Seats Available']} seats)")

    print("\n✈️ All Flight Availability:")
    for flight in flights:
        print(f"{flight['Flight ID']} to {flight['Destination']}: {flight['Seats Available']} seats @ ${flight['Price']} each")

# Initialize the flight system
initialize_flights()

# Example Usage
add_flight("F200", "San Francisco", 180, 450.0)  # Adding a new flight
search_flight("F101")  # Searching for an existing flight
book_seats("F101", 20)  # Booking 20 seats
cancel_seats("F101", 10)  # Canceling 10 seats
generate_report()  # Generating report

```
